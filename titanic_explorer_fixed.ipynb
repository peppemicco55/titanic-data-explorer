```python
df = pd.read_csv('titanic.csv')
# leggiamo il file csv
print(df.head())
print(df.info())
```

```python
df['HasCabin'] = df['Cabin'].notna().astype(int)
# aggiungiamo una colonna che indica se la cabina è presente o meno
print(df[["Cabin", "HasCabin"]].head(10))
```

```python
df.columns
```

```python
df["Age"].isna().sum()
# contiamo i valori mancanti nella colonna Age
```

```python
df["Age"].median()
# calcoliamo la mediana della colonna Age
```

```python
df["Age"] = df["Age"].fillna(27)
# sostituiamo i valori mancanti con la mediana
```

```python
df['Age'].isna().sum()
# verifichiamo che non ci siano più valori mancanti
```

```python
def fascia_eta(eta):
    if eta < 18:
        return "minorenne"
    elif eta < 30:
        return "giovane"
    elif eta < 50:
        return "adulto"
    else:
        return "senior"

df["FasciaEta"] = df["Age"].apply(fascia_eta)
# aggiungiamo una colonna con la fascia d'età
```

```python
print(df[["Age", "FasciaEta"]].head(10))
# mostriamo le prime 10 righe delle colonne Age e FasciaEta
```

```python
print(df["FasciaEta"].value_counts())
# contiamo quante persone ci sono in ogni fascia d'età
```

```python
df.to_csv("titanic_clean.csv", index=False)
# salviamo il dataframe pulito in un nuovo file csv
```

```python
print(pd.crosstab(df["Pclass"], df["HasCabin"]))
# creiamo una tabella di contingenza tra Pclass e HasCabin
```

```python
print(pd.crosstab(df["Pclass"], df["HasCabin"], normalize="index") * 100)
# calcoliamo le percentuali nella tabella di contingenza
```

```python
df.groupby("HasCabin")["Age"].mean()
# calcoliamo l'età media in base alla presenza della cabina
```

```python
df.groupby("HasCabin")["Age"].median()
# calcoliamo la mediana dell'età in base alla presenza della cabina
```

```python
df["FarePerPersona"] = df["Fare"] / (df["SibSp"] + df["Parch"] + 1)
# aggiungiamo una colonna con la tariffa per persona
print(df[["Fare", "SibSp", "Parch", "FarePerPersona"]].head(10))

```

```python
print(df.groupby("HasCabin")["FarePerPersona"].median())
# calcoliamo la mediana della tariffa per persona in base alla presenza della cabina
```

```python
(pd.crosstab(
    index=[df["FasciaEta"], df["Pclass"]],
    columns=df["Survived"]
))
# creiamo una tabella di contingenza tra FasciaEta, Pclass e Survived
```

```python
print(pd.crosstab(
    index=[df["FasciaEta"], df["Pclass"]],
    columns=df["Survived"],
    normalize="index"
) * 100)
# calcoliamo le percentuali nella tabella di contingenza tra FasciaEta, Pclass e Survived
```

```python
n_table = pd.crosstab(
    index=[df["FasciaEta"], df["Pclass"]],
    columns="N"
)
print(n_table)
# ragruppa la classe in base a FasciaEta e Pclass e conta il numero di occorrenze in ogni gruppo
```

```python
perc_table = pd.crosstab(
    index=[df["FasciaEta"], df["Pclass"]],
    columns=df["Survived"],
    normalize="index"
) * 100
# calcoliamo le percentuali nella tabella di contingenza tra FasciaEta, Pclass e Survived
```

```python
final_table = n_table.join(perc_table)
print(final_table)
# uniamo le due tabelle per avere il numero e le percentuali in un'unica tabella
```
